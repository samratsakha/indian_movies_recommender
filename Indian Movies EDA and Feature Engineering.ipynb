### Analysis on year

```python
year = list(df['Year'].unique())
for i in range(len(year)):
    year[i] = str(year[i])
year.sort()
```

```python
for i in year:
    print(i)
```

```python
years = list(df['Year'])
for i in range(len(years)):
    years[i] = str(years[i])
```

```python
def year_to_int(num):
    temp = ""
    count = 0
    for i in range(len(num)):
        if num[i].isdigit():
            temp += str(num[i])
            count+=1
            if count==4:
                break
                
    if count<4:
        temp = 0

    return temp 
```

```python
for i in range(len(years)):
    years[i] = year_to_int(years[i])
```

```python
for i in range(len(years)):
    years[i] = int(years[i])
```

```python
years_ready = np.array(years)
```

```python
df['Year'] = years_ready
```

### Analysis on Genre

```python
genre = list(df['Genre'])
```

```python
mylist = list()

count = 0
for i in genre:
    genre_list = i.split(",")
    for j in range(len(genre_list)):
        genre_list[j] = str(genre_list[j])
        genre_list[j] = genre_list[j].replace(" ","")
        genre_list[j] = genre_list[j].replace("-","0")
        mylist.append(genre_list[j])
```

```python
unique_genres = set(mylist)
```

```python
unique_genres
```

```python
unique_genres = list(unique_genres)
unique_genres.sort()
```

```python
unique_dict = dict()
for i in range(len(unique_genres)):
    unique_dict[unique_genres[i]] = i
```

```python
unique_dict
```

```python
genre_change = genre.copy()
len(genre_change)
```

```python
for i in range(len(genre_change)):
    genre_list = genre_change[i].split(",")
    val = ""
    
    for j in range(len(genre_list)):
        genre_list[j] = str(genre_list[j])
        genre_list[j] = genre_list[j].replace(" ","")
        genre_list[j] = genre_list[j].replace("-","0")
        
        temp = unique_dict[genre_list[j]]
        val += str(temp)
        
    genre_change[i] = val
```

```python
for i in range(len(genre_change)):
    print(genre[i],genre_change[i])
```

```python
for i in range(len(genre_change)):
    genre_change[i] = int(genre_change[i])
```

```python
df['Genre'] = np.array(genre_change)
```
